# 다양한 그래프 활용

한글이 포함된 경우 깨지는 현상이 나타난다. 이를 위해 따로 설정을 해야한다.

```python
from matplotlib import pyplot as plt, font_manager
import numpy as np
import math

# 한글 폰트 사용하기 위해 설정
# 윈도 10에 있는 기본 폰트 파일을 사용하도록 설정하기

import matplotlib.font_manager as fm
import matplotlib

font_loacation = r"C:/Windows/Fonts/H2GTRM.TTF" #윈도우에 있는 폰트 파일의 위치를 지정합니다.
font_name = fm.FontProperties(fname = font_loacation, size = 50).get_name() #폰트 파일의 설정에서 이름정보를 가져 조회합니다.
matplotlib.rc("font", family = font_name) #폰트파일의 이름을 matplotlib에 설정합니다.
```

```python
# 샘플데이터 - 사원별 월별 실적을 랜덤으로 생성함
np.random.seed(0)
hong = np.ceil(np.random.rand(12) * 100)
kang = np.ceil(np.random.rand(12) * 100)
date = np.arange(1, 13)
```

```python
# 라인그래프 설정
plt.style.use("ggplot")
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(date, hong, label = "홍길동")
ax.plot(date, kang, label = "강감찬")

# 축 제목과 범례 그리기
ax.set_title("영업 사원별 월별 실적")
ax.set_ylabel("실적(단위: 원)")
ax.set_xlabel("월")
ax.legend()
```
